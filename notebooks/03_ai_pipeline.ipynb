```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)

print("Libraries Imported Successfully")
```

```python
df = pd.read_csv("../data/processed/clean_customer_support_dataset.csv")

print(df.shape)

df.head()
```

```python
issue_df = df[
    [
        "initial_message",
        "issue_type"
    ]
].copy()
```

```python
issue_df.columns = [
    "text",
    "label"
]
```

```python
issue_df = issue_df.dropna()
```

```python
issue_df = issue_df.sample(
    frac=1,
    random_state=42
).reset_index(drop=True)
```

```python
issue_df.head()
```

```python
priority_df = df[
    [
        "initial_message",
        "priority"
    ]
].copy()
```

```python
priority_df.columns = [
    "text",
    "label"
]
```

```python
priority_df = priority_df.dropna()
```

```python
priority_df = priority_df.sample(
    frac=1,
    random_state=42
).reset_index(drop=True)
```

```python
priority_df.head()
```

```python
regression_df = df[
    df["is_resolved"] == 1
].copy()
```

```python
regression_df = regression_df[
    [
        "issue_type",
        "priority",
        "customer_segment",
        "channel",
        "platform",
        "region",
        "sla_plan",
        "reopened",
        "has_attachment",
        "message_length",
        "word_count",
        "resolution_time_hours"
    ]
]
```

```python
print(regression_df.shape)

regression_df.head()
```

```python
retrieval_df = df[
    [
        "initial_message",
        "agent_first_reply",
        "resolution_summary",
        "issue_type",
        "priority"
    ]
].copy()
```

```python
retrieval_df = retrieval_df.dropna(
    subset=["resolution_summary"]
)
```

```python
retrieval_df = retrieval_df.reset_index(drop=True)
```

```python
retrieval_df.head()
```

```python
issue_df.to_csv(
    "../data/processed/issue_classification.csv",
    index=False
)

priority_df.to_csv(
    "../data/processed/priority_prediction.csv",
    index=False
)

regression_df.to_csv(
    "../data/processed/regression_dataset.csv",
    index=False
)

retrieval_df.to_csv(
    "../data/processed/retrieval_dataset.csv",
    index=False
)

print("All AI datasets saved successfully!")
```
