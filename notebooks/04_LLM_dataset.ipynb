```python
import os
import pandas as pd

# Load unique issue dataset
issue_df = pd.read_csv("../data/processed/issue_classification.csv")

os.makedirs("../prompts", exist_ok=True)

labels = sorted(issue_df["label"].unique())

print(labels)
```

```python
for label in labels:

    examples = (
        issue_df[issue_df["label"] == label]["text"]
        .drop_duplicates()
        .tolist()
    )

    prompt = f"""
You are helping create a high-quality NLP dataset.

Issue Category:
{label}

Below are all the unique examples available in my dataset.

Study them carefully.

Understand the issue category.

DO NOT COPY THEM.

DO NOT PARAPHRASE THEM.

Generate exactly 50 NEW customer support tickets.

The new tickets must:

• belong to the SAME issue category

• describe DIFFERENT situations

• be realistic

• sound human

• include

- beginners
- enterprise users
- students
- developers
- elderly users
- frustrated users
- polite users

Mix

- short
- medium
- long

Return ONLY the tickets.

One ticket per line.

No numbering.

No markdown.

Examples:

"""

    for example in examples:

        prompt += f"- {example}\n"

    filename = f"../prompts/{label}.txt"

    with open(filename, "w", encoding="utf-8") as f:

        f.write(prompt)

print("Done!")
```

```python
import pandas as pd
import re
```

```python
DATA_FOLDER = ("../data/generated")
```

```python
files = {
    "account_cleaned_final.txt": "account_access",
    "bp_cleaned.txt": "billing_problem",
    "bug_cleaned.txt": "bug",
    "feature_cleaned.txt": "feature_request",
    "how_to_cleaned.txt": "how_to",
    "other_cleaned.txt": "other",
    "performance_cleaned.txt": "performance",
    "security_cleaned.txt": "security_concern",
}
```

```python
rows = []

for filename, label in files.items():

    filepath = os.path.join(DATA_FOLDER,filename)

    with open(filepath, "r", encoding="utf-8") as f:

        for line in f:

            line = line.strip()

            if not line:
                continue

            # Remove numbering like:
            # 1.
            # 24.
            # 103.

            if ". " in line:
                first, rest = line.split(". ", 1)

                if first.isdigit():
                    line = rest

            rows.append({
                "text": line,
                "label": label
            })

df = pd.DataFrame(rows)
```

```python
print(df.shape)

print()

print(df.head())
```

```python
print(df["label"].value_counts())
```

```python
print("Before:", len(df))

df = df.drop_duplicates(subset="text")

print("After :", len(df))
```

```python
df = df.sample(
    frac=1,
    random_state=42
).reset_index(drop=True)
```

```python
OUTPUT_PATH = "../data/processed/issue_classification.csv"

df.to_csv(
    OUTPUT_PATH,
    index=False
)

print("Dataset saved successfully!")

print(df.shape)
```
